```python
df_cian = pd.read_csv('output\data29_01_2017__21_29_12ADD.csv',sep=';',encoding='cp1251')
```

```python
def del_from_str(string,to_del):
    return string.replace(to_del,'')
def street_mask(street_name):
    to_del_list = ['ул.','пер.','просп.','наб.','ш.',' ул','бульв.','проезд ']
    for to_del in to_del_list:
        street_name = del_from_str(street_name,to_del)
    return street_name
```

```python
ADDRESS_LIST_CIAN = []
#for i in range(900,950):
for i in tqdm(range(len(df_cian))):
    address = df_cian['Address'].iloc[i]
    #print(address)
    address_in_list = address.split(',')
    street = street_mask(address_in_list[-2])
    street = del_from_str(street,' ')
    num = address_in_list[-1]
    num = del_from_str(num,' ')
    for add_part in address_in_list:
        #print(add_part)
        if 'район' in add_part:
            rayon = add_part.split(' ')[-1]
            #print(rayon)
            rayon = del_from_str(rayon,' ')
    ADDRESS = street + '_' + num
    ADDRESS_LIST_CIAN.append(ADDRESS)
    #print(ADDRESS)
```
